```python
# load dataset
import pandas as pd
df = pd.read_csv('data_tutorial.csv')
```

```python
# divide dataset into medical specialty and symptoms dataset

datasetY = df.iloc[:,-9:] # medical specialty data

ms = list(datasetY.columns) # list of medical specialty
datasetX = df.drop(list(datasetY.columns), axis = 1) # symptoms & comorbidities data
symp = list(datasetX.columns) # list of symptoms
datasetY = datasetY.to_numpy() # to array
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
import numpy as np
```

```python
# building the MLP model
def baseline_model():
    model = Sequential()
    model.add(Dense(240, input_dim=26, activation='relu', kernel_initializer='he_uniform'))
    model.add(Dense(180, activation='relu', kernel_initializer='he_uniform'))
    model.add(Dense(120, activation='relu', kernel_initializer='he_uniform'))
    model.add(Dense(60, activation='relu', kernel_initializer='he_uniform'))
    model.add(Dense(9, activation='softmax'))
    
    model.compile(loss='categorical_hinge', optimizer='adam', metrics=['accuracy'])

    return model


estimator = KerasClassifier(build_fn=baseline_model, epochs=40, batch_size=32, verbose=1)
```

```python
# leave one out cross validation to see the model performance
loo = LeaveOneOut()

predicted = cross_val_predict(estimator, datasetX, datasetY, cv=loo) # data predicted by loocv
actual = np.where(datasetY == 1)[1] # real data
```

```python
# precision, recall, f1, etc (comparison of actual vs. predicted data)
from sklearn.metrics import classification_report

print(classification_report(actual,predicted, target_names = ['cardio', 'inter', 'neuro', 'surg', 'ENT', 'derma', 'ophthal', 'pulmo', 'ortho'], zero_division=0))
```

```python
model = baseline_model()

# train model with all of the dataset
model.fit(datasetX, datasetY, epochs=40, batch_size=32,
          shuffle=True,
          verbose=1)

# save the model
model.save('chatbot_model1')
```
